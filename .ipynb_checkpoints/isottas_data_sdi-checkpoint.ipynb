```python
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.io as sio
import pandas as pd
import pygsp
```

```python


def fancyFig(ax):
    ax.tick_params(width=3, length=8)
    ax.spines['top'].set_linewidth(0)
    ax.spines['bottom'].set_linewidth(3)
    ax.spines['left'].set_linewidth(3)
    ax.spines['right'].set_linewidth(0)


def plotShaded(x, y, axis, color, label, ax, ls='-'):
    ax.plot(x, np.median(y, axis=axis), color=color, lw=2, label=label, ls=ls)
    ax.fill_between(x, *np.percentile(y, (5, 95), axis=axis), color=color,
                    alpha=.4)
    return ax


def plotShadedNorm(x, y, axis, color, label, ax, ls='-'):
    ax.plot(x, np.mean(y, axis=axis), color=color, lw=2, label=label, ls=ls)
    ax.fill_between(x, *[[1], [-1]]*np.std(y, axis=axis) +
                    np.mean(y, axis=axis), color=color, alpha=.4)
    return ax

```

```python
# Load ROI atlas info
scale=1
sc_dir = '/Volumes/SSD-Bercows/Joan/research/0_data/'
roifname = os.path.join(sc_dir, 'Lausanne2008_Yeo7RSNs.xlsx')
roidata = pd.read_excel(roifname, sheet_name='SCALE {}'.format(scale))
#cort = np.where(roidata['Structure'] == 'cort')[0]
cort = np.concatenate((np.arange(34),np.arange(39,76),np.arange(80,81)))
sc_fname = os.path.join(sc_dir, 'SC_betzel',
                        'SC_num_betzel_scale_{}.mat'.format(scale))
sc_file = scipy.io.loadmat(sc_fname)['num'][cort][:, cort]
sc_file[sc_file != 0] = np.log(sc_file[sc_file != 0])

sc_file[np.isnan(sc_file)] = 0
sc_file[np.isinf(sc_file)] = 0
sc = pygsp.graphs.Graph(sc_file,lap_type='normalized')
sc.compute_fourier_basis()
```

```python
import tqdm
def sdi(sc, data):
    nsub = data.shape[0]
    
    power = np.zeros((sc.N, data.shape[2], nsub))
    evoked_hat = np.zeros((sc.N,  data.shape[2], nsub))

    #for s in tqdm.tqdm(range(nsub)):
    for s in range(nsub):
        epochs = data[s]
        epochs_hat = sc.gft(epochs)
        evoked_hat[:, :, s] = epochs_hat.copy()

        # power of the signal in the graph
        power[:, :, s] = abs(epochs_hat) ** 2

    # mean across time
    psd = np.mean(power, 1)

    # mean across subjects
    m_psd = np.mean(psd, 1)

    # area under the curve
    auc_tot = np.trapz(m_psd)
    i = 0
    auc = 0
    while auc < (auc_tot / 2):
        i += 1
        auc = np.trapz(m_psd[:i])
    # cut-off frequency
    nn = i - 1
    vhigh = np.zeros_like(sc.U)
    vlow = np.zeros_like(sc.U)
    vhigh[:, nn:] = sc.U[:, nn:]
    vlow[:, :nn] = sc.U[:, :nn]

    # coupling / decoupling

    x_c = np.zeros((sc.N, data.shape[2], nsub))
    x_d = np.zeros((sc.N, data.shape[2], nsub))
    n_c = np.zeros((sc.N, nsub))
    n_d = np.zeros((sc.N, nsub))

    #for s in tqdm.tqdm(range(nsub)):
    for s in range(nsub):
        x_c[:, :, s] = vlow @ evoked_hat[:, :, s]
        x_d[:, :, s] = vhigh @ evoked_hat[:, :, s]

    return psd, nn, x_d, x_c

```

```python
path2data = '/Volumes/SSD-Bercows/Joan/research/0_data/DS_IEDs/derivatives/source_modelling/'
subject_list = [elem for elem in os.listdir(path2data) if elem.startswith('sub')]
```

```python
fs = 250
time = np.linspace(0,1,fs)
time_analysis = np.where((time>=0.3) & (time<=0.7))[0]
```

```python

d_avg_timeseries = np.zeros((len(subject_list),len(time_analysis)))
c_avg_timeseries = np.zeros((len(subject_list),len(time_analysis)))

d_avg_timeseries_norm = np.zeros((len(subject_list),len(time_analysis)))
c_avg_timeseries_norm = np.zeros((len(subject_list),len(time_analysis)))

d_avg_timeseries_norm2 = np.zeros((len(subject_list),len(time_analysis)))
c_avg_timeseries_norm2 = np.zeros((len(subject_list),len(time_analysis)))


for s,sub in enumerate(subject_list):
    if s == 9:
        file = [elem for elem in os.listdir(os.path.join(path2data,sub,'ses-preop','eeg'))
                if (elem.startswith('sub') & elem.endswith('_roi_time_series.mat'))][0]

        eeg = sio.loadmat(os.path.join(path2data,sub,'ses-preop','eeg',file))    
        eegdata = eeg['roi'][0,0][-1][0]
        n_trials = len(eegdata)
        eeg_trials = np.zeros((n_trials,72,len(time_analysis)))

        power_in_time = np.zeros((n_trials,len(time_analysis)))

        for t in range(n_trials):
            eeg_trials[t] = eegdata[t][:,time_analysis]
            eeg_trials[t] /= np.std(eeg_trials[t].flatten()) - \
                np.mean(eeg_trials[t].flatten())
            power_in_time[t] = np.linalg.norm(eeg_trials[t],ord=2,axis=0)

        #power_in_time /= np.sum(power_in_time,axis=1,keepdims=True)
        power2 = power_in_time.copy()
        power2 /= np.sum(power2,axis=1,keepdims=True)
        psd, nn, x_d, x_c = sdi(sc,eeg_trials)

        d_avg_timeseries[s] = np.linalg.norm(x_d,2,0).mean(1)
        c_avg_timeseries[s] = np.linalg.norm(x_c,2,0).mean(1)

        d_avg_timeseries_norm[s] = (np.linalg.norm(x_d,2,0)/power_in_time.T).mean(1)
        c_avg_timeseries_norm[s] = (np.linalg.norm(x_c,2,0)/power_in_time.T).mean(1)
        
        d_avg_timeseries_norm2[s] = (np.linalg.norm(x_d,2,0)/power2.T).mean(1)
        c_avg_timeseries_norm2[s] = (np.linalg.norm(x_c,2,0)/power2.T).mean(1)
        

        c_avg_timeseries_norm2[s] = (np.linalg.norm(x_c,2,0)/power2.T).mean(1)
        

        print(np.trapz(psd.mean(1)[:nn]),np.trapz(psd.mean(1)[nn:]),nn)
```
